```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import sympy as sp
import math
x,y,z = sp.symbols('x,y,z')
```

```python
def f(x):
    return x**2

x  = np.linspace(0,10, 3000)
y = f(x)

plt.plot(x,y)
plt.title("Y = X^2")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.fill_between(x,f(x) , where = [(x>3) and (x < 6) for x in x], color = 'r' )

plt.show()
```

```python
x**2
```

```python
sp.integrate(x**2, (x, 3 , 6))
```

```python
def f(x):
    return np.sin(x)

x = np.linspace(0,np.pi,3000)

plt.plot(x,f(x))
plt.title("Y = sin(x)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.fill_between(x,f(x) , where = [(x>0) and (x <math.pi/2) for x in x], color = 'k' )
plt.show()
```

```python

sp.integrate(sp.sin(x) ,(x,0, math.pi/2))
```

```python
def f(x,a):
    
    return a*(math.e**x)

x = np.linspace(10,100,num = 3000)
y = f(x,1)

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("y = e^x")
#plt.fill_between(x,f , where = [(x>0) and (x < 100) for x in x], color = 'g' )
plt.show()
```

```python
def f(x):
    return np.log(x**2)

x = np.linspace(-10,10, 100)

y = f(x)

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("y = (x**2)/x**2 + 1")
plt.fill_between(x,y , where = [(x>1) and (x < 10) for x in x], color = 'g' )
plt.show()
```

```python
sp.integrate(sp.log(x) ,(x,1,10))
```

```python
def f(x,a):
    return a*np.arcsin(x)
x = np.linspace((-math.pi)/2,(math.pi)/2,3000)
y = f(x,1)

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("y = arcsin(x)" )
plt.fill_between(x,y , where = [(x>0) and (x < (math.pi)/2 ) for x in x], color = 'b' )
plt.show()
```

```python
sp.integrate(sp.asin(x) ,( x , 0 ,1 ))
```

```python
def f(x):
    return np.sqrt(1/(1-x**2))

x = np.linspace(-10,10,3000)
y = f(x)

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("y = sqrt(1/(1-x^2)" )
plt.fill_between(x,y , where = [(x>0) and (x < 1 ) for x in x], color = 'k' )
plt.show()
```

```python
sp.integrate(sp.sqrt(1/(1-x**2)) , (x ,0 ,1))
```
