```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import math
x,y,z = sp.symbols('x y z')
```

```python
sp.sin(x) + sp.cos(x)
```

```python
def f(x):
    return np.sin(x) + np.cos(x)

x = np.linspace(0,10,3000)
y = f(x)

plt.plot(x,f(x))
plt.title("Y = sin(x) + cos(x)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.fill_between(x,f(x) , where = [(x>0) and (x < 2*math.pi) for x in x] , color = 'k')

plt.show()
```

```python
sp.integrate(sp.sin(x) + sp.cos(x) ,( x , 0 , math.pi ))
```

```python
sp.sqrt(x-1)
```

```python
def f(x):
    return np.sqrt(x-1)

x = np.linspace(1,10,3000)
y = f(x)

plt.plot(x,f(x))
plt.title("Y = sqrt(x-1)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.fill_between(x,f(x) , where = [(x>0) and (x < 9) for x in x] , color = 'r')

plt.show()
```

```python
sp.integrate(sp.sqrt(x-1) , (x,1,9))
```

```python
def f(x):
    return np.tan(x)

x = np.linspace((-2*math.pi),( 2*math.pi) ,num = 100)
y = f(x)

plt.plot(x,y)
plt.title("Y = tan(x) ")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.fill_between(x, y , where = [(x>0) and (x < math.pi) for x in x] , color = 'r')

plt.show()
```
